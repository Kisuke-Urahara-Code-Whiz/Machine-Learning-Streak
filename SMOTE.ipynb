### Why smote?

Basically, in normal cases, what happens is that traditional upsampling adds similar data points in the same place, which isnt the best option as there is less variance in the upsampled data, as we are duplicatiing the present datapoints at the same place <br>

SMOTE is diff, wt it normally does is, if looked in a graph, it first draws a line between two data points and add more data points along the line thus adding more data points with more variance and not stuck at the same place, which is a better method of upsampling data. 

## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification

## this helps us to make a binary class or multi class classification dataset
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                   weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```
